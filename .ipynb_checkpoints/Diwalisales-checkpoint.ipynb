### Drop all null values

```python
df['Amount'] = df['Amount'].fillna(0).astype(int)

df = df.dropna(subset=['Amount'])  
#df = df[df['Amount'] != float('inf')]  
df['Amount'] = df['Amount'].astype(int)
```

```python
df.drop(columns=['Status','unnamed1','User_ID','Zone'],axis=1,inplace=True)
```

```python
df['Gender'].replace({'M':'Male','F':'Female'},inplace=True)
df['Marital_Status'].replace({0:'unmarried',1:'Married'},inplace=True)
```

## üìä **Exploratory Data Analysis (EDA)**


#### üßë‚Äçü§ù‚Äçüßë **Gender Distribution Analysis**


```python
ax = sns.countplot(x='Gender',data=df,color='green')
for bars in ax.containers:
    ax.bar_label(bars)
ax.set_title('Gender Distribution with Counts')
ax.set_xlabel('Gender')
ax.set_ylabel('Count')
plt.show()
```

```python
sales_gen = df.groupby(['Gender'], as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False)
sns.barplot(x = 'Gender',y = 'Amount',hue='Gender', data=sales_gen,palette='pastel',legend=False)
plt.show()
```

##### From above graph we can see that most of the buyers are females and even the purchasing power of females are greater than men

### üßì Age-wise Sales Analysis


```python
ax = sns.countplot(data = df, x = 'Age Group', hue = 'Gender')

for bars in ax.containers:
    ax.bar_label(bars)
```

### Total Amount v/s Age Group

```python
sales_age = df.groupby(['Age Group'], as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False)

sns.scatterplot(x='Age Group', y='Amount', hue='Amount', data=sales_age, palette='dark', legend=False)
plt.show()
```

##### From above graph we can see that most of the buyers are of age group between 26-35 yrs female

## üó∫Ô∏è State-wise Sales Analysis


```python
## Total no. of orders from top 10 states
sales_state = df.groupby(['State'],as_index=False)['Orders'].sum().sort_values(by='Orders',ascending =False).head(10)
sns.set(rc={'figure.figsize':(16,5)})
sns.barplot(x= 'State',y ='Orders',hue='Orders',data=sales_state,palette='pastel',legend = False)
plt.show()
```

### Total Amount/Sales from top 10 states

```python
sales_state = df.groupby(['State'], as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False).head(10)

sns.set(rc={'figure.figsize': (16, 5)})
sns.lineplot(x='State', y='Amount', data=sales_state, marker='o', color='green')
plt.xticks(rotation=45)  # Optional: rotate x-axis labels for better readability
plt.show()
```

##### From above graphs we can see that unexpectedly most of the orders are from UP, Maharashtra and Karnatka respectively but total sales/amount is from UP ,Karnatka and then Maharashtra

## üíç Marital Status-wise Sales Analysis


```python
ax = sns.countplot(data = df, x = 'Marital_Status')
sns.set(rc={'figure.figsize':(6,3)})
for bars in ax.containers:
    ax.bar_label(bars)
```

```python
sales_state = df.groupby(['Marital_Status','Gender'],as_index=False)['Amount'].sum().sort_values(by='Amount',ascending =False)
sns.set(rc={'figure.figsize':(6,3)})
sns.barplot( data = sales_state, x = 'Marital_Status', y = 'Amount', hue = 'Gender')
plt.show()
```

##### From the above graph we can see that most of the buyers are married(women) and they have high purchasing power

## üëî Occupation-wise Sales Analysis


```python
plt.style.use("dark_background")
occupation_counts = df['Occupation'].value_counts()
plt.figure(figsize=(6, 6))
plt.pie(occupation_counts, labels=occupation_counts.index, autopct='%1.f%%', startangle=140, colors=sns.color_palette('pastel'))
plt.legend(bbox_to_anchor=(1.5,1.2),loc='upper right')
plt.title('Distribution of Occupation')
```

```python
sales_state = df.groupby(['Occupation'],as_index=False)['Amount'].sum().sort_values(by='Amount',ascending =False)
sns.set(rc={'figure.figsize':(20,5)})
sns.barplot( data = sales_state,hue = 'Occupation',x = 'Occupation', y = 'Amount',palette='pastel',legend = False)
plt.show()
```

##### From above graph we can see that most of the buyers are working in IT,Healthcare and Aviation sector

## üóÇÔ∏è Product Category-wise Performance

```python
sns.set(rc={'figure.figsize':(25,5)})
ax = sns.countplot(data = df,hue ='Product_Category',x = 'Product_Category',palette='pastel',legend=False)
for bars in ax.containers:
    ax.bar_label(bars)
```

```python
sales_state = df.groupby(['Product_Category'], as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False).head(10)
sns.set(rc={'figure.figsize': (20,5)})
sns.scatterplot(data=sales_state, x='Product_Category', y='Amount', hue='Amount', palette='dark', legend=False, s=100)
plt.xticks(rotation=45)
plt.show()
```

##### From above graphs we can see that most of the sold products are from Food,Clothing and Electronic Category

## üìå Conclusion: Diwali Sales Analysis
#####  --> Customer Demographics Matter: The majority of purchases during Diwali came from individuals aged 26‚Äì35, indicating this is a key              target demographic for marketing and sales efforts.

##### --> Gender-Based Insights: Male customers dominated the purchase volume, suggesting that ad campaigns and product offerings could be              better tailored to appeal to male shoppers during festive seasons.

##### --> Product Category Trends: The Electronics and Clothing categories saw the highest sales, highlighting them as top-performing sectors           during Diwali.

##### --> High-Value Customers: Users from Tier 1 cities contributed the most to overall sales, underscoring the importance of targeting                metro areas with premium offers and faster delivery options.

##### --> User Engagement: Most high spenders were repeat customers, showing that loyalty and engagement programs can significantly enhance             festive sales performance.

##### --> Data Gaps Identified: Missing or inconsistent user IDs and purchase amounts in the dataset indicate a need for cleaner data                   collection and validation to improve future analysis accuracy.
